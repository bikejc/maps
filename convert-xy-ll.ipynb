```python
with open('public/Proposed Full Network.geojson', 'r') as f:
    gj = json.load(f)
gj
```

```python
from pyproj import Proj, transform

inProj = Proj(init='epsg:3424')
outProj = Proj(init='epsg:4326')
```

```python
features = gj['features']
for idx, feature in enumerate(features):
    properties = feature['properties']
    Street = properties['Street']
    geometry = feature['geometry']
    if geometry['type'] != 'LineString':
        raise ValueError(f'{idx}/{Street}: {geometry["type"]}')
    geometry['coordinates'] = [ transform(inProj, outProj, x, y) for x, y in geometry['coordinates'] ]
```

```python
gj['crs']['properties']['name'] = 'urn:ogc:def:crs:EPSG::4326'
gj
```

```python
with open('public/Proposed Full Network.geojson', 'w') as f:
    json.dump(gj, f, indent=4)
```
